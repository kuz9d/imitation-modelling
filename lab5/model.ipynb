```python
import networkx as nx
import matplotlib.pyplot as plt
import random

N = 1000 # количество узлов
m = 2 # плотность связей
G = nx.barabasi_albert_graph(N, m)

states = {node: 'S' for node in G.nodes()}
initial_infected = random.choice(list(G.nodes()))
states[initial_infected] = 'I'

beta = 0.15  # вероятность заражения
gamma = 0.05  # вероятность выздоровления
steps = 50

S = [N - 1]
I = [1]
R = [0]

for _ in range(steps):
    new_infections = []
    new_recovered = []

    for node in G.nodes():
        if states[node] == 'I':
            for neighbor in G.neighbors(node):
                if states[neighbor] == 'S' and random.random() < beta:
                    new_infections.append(neighbor)
            if random.random() < gamma:
                new_recovered.append(node)

    for node in new_infections:
        states[node] = 'I'
    for node in new_recovered:
        states[node] = 'R'

    S.append(sum(1 for s in states.values() if s == 'S'))
    I.append(sum(1 for s in states.values() if s == 'I'))
    R.append(sum(1 for s in states.values() if s == 'R'))

plt.figure(figsize=(10, 6))
plt.plot(S, label='Susceptible')
plt.plot(I, label='Infected')
plt.plot(R, label='Recovered')
plt.xlabel('Шаги')
plt.ylabel('Количество узлов')
plt.legend()
plt.title('Динамика распространения вируса (SIR модель)')
plt.show()
```

```python
# Параметры для сравнения
m_values = [1, 2, 3, 5]
beta = 0.15
gamma = 0.05
steps = 50
initial_infected = 1

plt.figure(figsize=(12, 8))
for m in m_values:
    G = nx.barabasi_albert_graph(N, m)
    states = {node: 'S' for node in G.nodes()}
    initial_nodes = random.sample(list(G.nodes()), initial_infected)
    for node in initial_nodes:
        states[node] = 'I'

    I_counts = [initial_infected]
    for _ in range(steps):
        new_infections = []
        new_recovered = []
        for node in G.nodes():
            if states[node] == 'I':
                for neighbor in G.neighbors(node):
                    if states[neighbor] == 'S' and random.random() < beta:
                        new_infections.append(neighbor)
                if random.random() < gamma:
                    new_recovered.append(node)

        for node in new_infections:
            states[node] = 'I'
        for node in new_recovered:
            states[node] = 'R'

        I_counts.append(sum(1 for s in states.values() if s == 'I'))

    plt.plot(I_counts, label=f'm={m}')

plt.xlabel('Шаги')
plt.ylabel('Количество инфицированных')
plt.title('Влияние плотности связей (m) на распространение')
plt.legend()
plt.show()
```
